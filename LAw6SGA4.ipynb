```python
import numpy as np
import numpy.linalg as la
import math
import matplotlib.pyplot as plt

A = np.array([[0.73,0.12,0.1], [0.15,0.73,0.2],[0.12,0.15,0.7]])
p1 = np.array([[9400],[500],[100]])
H = []
M = []
D = []
time = []


for i in range(1, 25):
    time.append(i)
    p = A@p1
    H.append(p[0][0])
    M.append(p[1][0])
    D.append(p[2][0])
    p1 = p

plt.xlabel('Time in days')
plt.ylabel('Number of people by category')
plt.plot(time, H, label="Healthy")
plt.plot(time, M, label='Mild')
plt.plot(time, D, label='Difficult')
plt.legend()
    
```
